```python
import pandas as pd
import matplotlib.pyplot as plt

titanic_data = pd.read_csv(r'Downloads/titanic/train.csv')

# Bar chart
titanic_data['Sex'].value_counts().plot(kind='bar', color=['blue', 'orange'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

# Histogram
titanic_data['Age'].dropna().plot(kind='hist', bins=10, color='green', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
